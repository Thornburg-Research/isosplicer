```python
gtfDir = './GCF_000001405.38/'
outDir = '/Data2/zane/human_genome/GCF_000001405.38/'
```

```python
import pyranges as pr
from matplotlib import pyplot as plt
import numpy as np

import os
```

```python
try:
    os.mkdir(outDir)
except:
    print('Failed to make outPut directory')
```

```python
geneList = []

for chrID in range(1,25):

    gtfFile = f"{gtfDir}chr{chrID}.gtf"
    print(gtfFile)
    
    gtf = pr.read_gtf(gtfFile)
    
    gtfdf = gtf.df

    geneIDs = set(gtfdf['gene_id'].values)
    print(len(geneIDs))

    for geneID in geneIDs:

        if '/' in geneID:
            print(geneID)
            continue
        
        geneDF = gtfdf.loc[(gtfdf['Feature'] == 'exon') | (gtfdf['Feature'] == 'gene')].loc[gtfdf['Source'].str.contains('BestRefSeq')].loc[gtfdf['gene_id']==geneID]
    
        if geneDF.shape[0]>2:

            if geneID in geneList:
                listIdx = geneList.index(geneID)
                print(geneID, geneList[listIdx])

            geneList.append(geneID)
            
            geneDF.to_excel(f'{outDir}{geneID}.xlsx', index=False)
```

```python
with open(f"{outDir}geneNames.txt", 'w') as f:
    for line in geneList:
        f.write(f"{line}\n")
```

```python
geneCounts = []

for chrID in range(1,25):

    gtfFile = f"{gtfDir}chr{chrID}.gtf"
    print(gtfFile)
    
    gtf = pr.read_gtf(gtfFile)
    
    gtfdf = gtf.df

    geneIDs = set(gtfdf['gene_id'].values)
    print(len(geneIDs))

    geneCounts.append(len(geneIDs))
```

```python
plt.rcParams.update({'font.size': 12})
plt.rcParams.update({"font.family": 'sans-serif'})

plt.rcParams['figure.dpi'] = 300

# plt.figure(figsize=(3.3,2))
fig = plt.figure(figsize=(10,4), layout='constrained')

ax = plt.gca()

tick_labels = []

chrCount = [i for i in range(1,25)]

for i in range(1,25):
    if i == 23:
        tick_labels.append('X')
    elif i == 24:
        tick_labels.append('Y')
    else:
        tick_labels.append(i)

    rect = ax.bar(i, geneCounts[i-1], color='green', edgecolor='k')
    ax.bar_label(rect, padding=3, rotation=45)

ax.set_xticks(chrCount,tick_labels)

ax.set_ylim(0,max(geneCounts)*1.2)

ax.plot(np.nan, np.nan, '-', color='none', label=f'NCBI Genome Entry: GRCh38.p12 / GCF_000001405.38')

ax.set_xlabel('Chromosome')
ax.set_ylabel('Gene Count')

ax.legend(frameon=False, loc='upper right')

plt.savefig('./chromosome_gene_counts.png')

plt.show()
```
